## Handle weather missing data

```python
from sklearn.impute import SimpleImputer

# Fill missing wind direction with most common value (mode)
mode_wind_dir = metar_df["wind_dir_degrees"].mode()[0]
metar_df["wind_dir_degrees"] = metar_df["wind_dir_degrees"].fillna(mode_wind_dir)

# Use median imputation for numerical weather data
num_cols = ["temperature_c", "dewpoint_c", "altimeter_hpa"]
imputer = SimpleImputer(strategy="median") 
metar_df[num_cols] = imputer.fit_transform(metar_df[num_cols])

# Verify that missing values
print("Missing Values After Handling:")
print(metar_df.isnull().sum())
```

1. wind_dir_degrees (Wind Direction) → Mode
Why? Wind direction is categorical (e.g., 0°–360°) and does not have a meaningful average.
Solution: We use mode (most common wind direction) because it represents the most likely value.
2. temperature_c, dewpoint_c, altimeter_hpa → Median
Why? These are numerical values, but weather measurements can have extreme values (outliers).
Solution: We use median (middle value) instead of mean because median is less affected by extreme highs or lows.

##  Undersampling (Reducing Majority Class)
### We balanced the dataset by downsampling the majority class(on-time flights).

```python
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = san_ot_df[san_ot_df["DEP_DEL15"] == 0]
df_minority = san_ot_df[san_ot_df["DEP_DEL15"] == 1]

# Downsample majority class (match minority class count)
df_majority_downsampled = resample(df_majority, 
                                   replace=False,  # Sample without replacement
                                   n_samples=len(df_minority),  # Match minority class count
                                   random_state=42) 

# Combine minority class with downsampled majority class
balanced_df = pd.concat([df_majority_downsampled, df_minority])

# Check new class distribution
print("Balanced Flight Delay Distribution:")
print(balanced_df["DEP_DEL15"].value_counts(normalize=True) * 100)
```

## Check if there is findings on seasonal bias

```python
import matplotlib.pyplot as plt

# Extract month from flight data
san_ot_df["Month"] = san_ot_df["FL_DATE"].dt.month

# Group by month and calculate delay percentage
monthly_delays = san_ot_df.groupby("Month")["DEP_DEL15"].mean() * 100

# Plot the seasonal delay trend
plt.figure(figsize=(10, 5))
monthly_delays.plot(kind="bar", color="skyblue")
plt.title("Seasonal Delay Trends")
plt.xlabel("Month")
plt.ylabel("Percentage of Delayed Flights")
plt.xticks(ticks=range(0,12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

# Conclusion on Findings

### Winter Months (Dec–Feb) → Moderate Delays

January and February have above-average delays, possibly due to rain, fog, and winter storms.
December has lower delays, which is unexpected given holiday travel but might indicate improved scheduling or fewer flights.
### Spring (March–May) → Mixed Trends

March shows one of the highest delay rates (~12%), possibly due to spring break travel congestion.
April and May have moderate delays.
### Summer (June–Aug) → Most Delays in June & July

June has the highest delay rate (~12.5%), likely due to increased travel demand and thunderstorms.
July also has high delays, while August sees a slight decline.

### Fall (Sept–Nov) → Steady Decrease

September and October have moderate delays.
November and December have the lowest delay percentages, which could indicate airlines improving scheduling before the holidays.

## What Can We do Based on These Findings?

We should include “Season” as a feature since delays clearly depend on the time of year.
December and November need a closer look—why are delays lower than expected? (maybe more cancellations) 
Weather might play a bigger role in summer delays—we should check wind speed, storms, and visibility data.

# Lets Add a Season Column 

```python
# Function to map months to seasons
def assign_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

# Add the season column based on flight month
san_ot_df["Season"] = san_ot_df["FL_DATE"].dt.month.apply(assign_season)

# Verify changes
print(san_ot_df[["FL_DATE", "Season"]].head())
```
